```python
import os
path = './results/FMErdf/object/'
files = [f for f in os.listdir(path) ]
```

```python
from bs4 import BeautifulSoup
```

```python
len(files)
```

```python
formulas = []
publication = []
persons = []
for obj in files:
    with open(path + obj, 'r') as f:
        data = f.read()
        xml = BeautifulSoup(data, 'xml')
        tp = xml.findChild('InformationObject').findChild("type").get('rdf:resource')
        if (tp == 'http://libmeta.ru/resource/Formula'):
            formulas.append(obj)
        elif (tp == 'http://libmeta.ru/resource/publication'):
            publication.append(obj)
        elif (tp == 'http://libmeta.ru/resource/person'):
            persons.append(obj)
```

```python
path_concept = './results/FMErdf/concept/'
files_concept = [f for f in os.listdir(path_concept) ]
```

```python
len(files_concept)
```

```python
from xml.dom import minidom

dom = minidom.Document()

root = dom.createElement('rdf:RDF')
root.setAttribute("xmlns:rdf", 'http://www.w3.org/1999/02/22-rdf-syntax-ns#')
root.setAttribute("xmlns", 'http://libmeta.ru/')


for obj in files_concept:

    with open(path_concept + obj, 'r') as f:
        data = f.read()
    print(path_concept + obj)
    xml = BeautifulSoup(data, 'xml')
    uri = xml.findChild('Concept').get('rdf:about')

    person = dom.createElement("Concept")
    person.setAttribute("rdf:about", uri)

    info_type = dom.createElement("thesaurus")
    info_type.setAttribute("rdf:resource", 'http://libmeta.ru/thesaurus/528624')
    person.appendChild(info_type)

    root.appendChild(person)




dom.appendChild(root)

f3 = open("ALL_CONCEPT_URIS_2.xml", "w")
f3.write(dom.toprettyxml())
f3.close()

```
