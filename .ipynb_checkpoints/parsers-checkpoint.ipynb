# Базовый парсер

Разбивает статьи на заголовки и текст (сшивая страницы).

Разбивка происходит в полуручном режиме, т.к. нет уверенности в формате заголовков.

В тексте ищутся фрагменты `" - "` и влево от них ищутся непрарывные последовательности слов, написанные капсом.

 - При удовлетворительном определении заголовка нажать `Enter` без дополнительного ввода.
 - При неправильном определении длины заголовка (начало не в том месте, где нужно) ввести верное количество слов.
 - Если предложенное место заголовком не является ввести `"n"`

В выводе переносы строк для удобства заменены на `"$"`

```python
from os import walk
import xml.etree.ElementTree as ET
from xml.dom import minidom
import re


############################ VARS ################################
pages_dir = "./matphys/rpages/"
exit_dir = "./matphys/"
exit_file = "FME.xml"
start_page = 1
end_page = 6
##################################################################


def prettify(elem):
    # Pretty-printed XML string for the Element.
    rough_string = ET.tostring(elem, 'utf-8')
    reparsed = minidom.parseString(rough_string)
    return reparsed.toprettyxml(indent="  ")
def xml_write():
    with open(exit_dir + exit_file, 'w') as f:
        f.write(prettify(root))


# Get filenames needed
filenames_raw = next(walk(pages_dir), (None, None, []))[2]  # [] if no file
filenames = []
for i in range(start_page, end_page + 1):
    for filename in filenames_raw:
        beginning = "rp-" + str(i) + "_"
        if filename[:len(beginning)] == beginning and filename[-4:] == ".mmd":
            filenames.append(filename)
            

# Check for existing xml
filenames_raw = next(walk(exit_dir), (None, None, []))[2]  # [] if no file
if not(exit_file in filenames_raw):
    root = ET.Element('data')
    xml_write()


# Parse existing xml (string parsing is needed to avoid extra newlines appearing)
exit_string = ''
with open(exit_dir + exit_file, 'r') as f:
    for i in f.readlines():
        exit_string += i[:-1]
root = ET.fromstring(exit_string)
# Remove empty tails and texts
root.tail = None
root.text = None
for i in root:
    i.tail = None
    i.text = None
    for j in i:
        j.tail = None


#article = ET.SubElement(root, 'article', {'n':str(len(root))})
#title = ET.SubElement(article, 'title')
#title.text = "test_title"
#text = ET.SubElement(article, 'text')
#text.text = "test_text"
#xml_write()


num = 0
pattern = re.compile
for filename in filenames:
    print()
    print("################################ " + filename + " ################################")
    with open(pages_dir + filename, 'r', encoding='utf8') as f:
        file = f.read()
    for i in range(len(file) - 2):
        if file[i:i+3] == ' - ':
            num += 1

            # Find title and 2 additional word on each side
            words = 0
            prev_ = i
            next_ = i + 3
            start_title = i
            end_title = i
            is_title = True
            while (is_title):
                prev_space = file.rfind(' ', 0, prev_)
                prev_space = 0 if prev_space == -1 else prev_space
                prev_nl = file.rfind('\n', 0, prev_)
                prev_nl = 0 if prev_nl == -1 else prev_nl
                prev_ = max(prev_nl, prev_space)
                word = file[0 if prev_ == 0 else prev_ + 1:start_title]
                is_title = False if prev_ == start_title or len(word) == 0 else True
                #print("\'", word, "\'", sep='')
                for letter in word:
                    is_title = False if re.match(r"[A-ZА-Я0-9]|[!#$%&'*+-.^_`|~:]", letter) == None else is_title
                if is_title:
                    start_title = prev_
                    words += 1
                #print(words, prev_, start_title, is_title)
            prev_space = file.rfind(' ', 0, prev_)
            prev_space = 0 if prev_space == -1 else prev_space
            prev_nl = file.rfind('\n', 0, prev_)
            prev_nl = 0 if prev_nl == -1 else prev_nl
            prev_ = max(prev_nl, prev_space)

            next_space = file.find(' ', next_, len(file))
            next_space = len(file) if next_space == -1 else next_space
            next_nl = file.find(' ', next_, len(file))
            next_nl = len(file) if next_nl == -1 else next_nl
            next_ = max(next_nl, next_space)
            next_space = file.find(' ', next_+1, len(file))
            next_space = len(file) if next_space == -1 else next_space
            next_nl = file.find(' ', next_, len(file))
            next_nl = len(file) if next_nl == -1 else next_nl
            next_ = max(next_nl, next_space)

            # Format terminal output
            out_str = file[0 if prev_ == 0 else prev_ + 1:next_]
            for j in range(len(out_str)):
                out_str = out_str[:j] + '$' + out_str[j+1:] if out_str[j] == '\n' else out_str
            print(str(num) + ')\n', out_str, sep='')
            print(' ' * (start_title - prev_) + '^' * (end_title - start_title - 1), end='    ')
            print("Words:", words) # , ';    ', file[start_title+1:end_title]
            #print(words, prev_, start_title, end_title)

            response = input()
            if response == '':
                print('Writing title: \"', file[start_title+1:end_title], '\"', sep='')
            elif response =='n':
                print('Not a title, skipping.')
            print()
```
